```sql
-- RESTORE JOAO PEDRO

/*
	Verificando títulos que estão com diferença e não deve fazer o update
	Ira gravar também o motivo de não ter feito o update
*/

USE dbSigaOpcaoPE


	UPDATE	tcba
		SET	StIdTituloNoAgenteDiferente			= CASE WHEN tcba.IdTituloNoAgente <> tc.IdTituloNoAgente THEN 1 ELSE 0 END,
			StDataVencimentoDiferente			= CASE WHEN tcba.DataVencimento <> tc.DataVencimento THEN 1 ELSE 0 END,
			StValorServicoDiferente				= CASE WHEN tcba.ValorServico <> tc.ValorServico THEN 1 ELSE 0 END,
			StValorDescontoNaoCondicionadoDiferente		= CASE WHEN tcba.ValorDescontoNaoCondicionado <> tc.ValorDescontoNaoCondicionado THEN 1 ELSE 0 END,
			StSomaValorDescontoCondicionadoDiferente	= CASE WHEN tcba.SomaValorDescontoCondicionado <> tdc.SomaValorDescontoCondicionado THEN 1 ELSE 0 END,
			StCPFCNPJDiferente				= CASE WHEN tcba.CPFCNPJ <> r.CPFCNPJ THEN 1 ELSE 0 END,
			StRealizarUpdate				= 0

		FROM	TbTituloCobrancaBoleto tcb
		INNER	JOIN VwTituloCobranca tc ON tc.IdTituloCobranca = tcb.IdTituloCobranca
		INNER	JOIN TbResponsavel r ON r.IdResponsavel = tc.IdResponsavel
		INNER	JOIN TbFormaRecebimento fr ON fr.IdFormaRecebimento = tc.IdFormaRecebimento
		OUTER	APPLY ( SELECT	SomaValorDescontoCondicionado = SUM(ValorDescontoCondicionado)
				FROM	VwTituloCobranca_DescontoCondicionado
				WHERE	IdTituloCobranca = tc.IdTituloCobranca ) tdc
		INNER	JOIN VPS7_TituloCobrancaBoleto.._CRM229504_OPCAO_TbTituloCobrancaBoleto_Anterior tcba ON tcba.IdTituloCobranca = tcb.IdTituloCobranca

		WHERE	fr.BOL_TipoCarteira = 'PJBank_1'
		AND	tc.SituacaoTituloCobranca = 'ABE'
		AND	tc.SituacaoCobrancaRegistrada IN ('RegistroConfirmado', 'RegistroPendente')
		AND	(
					tcba.IdTituloNoAgente			<> tc.IdTituloNoAgente
				OR	tcba.DataVencimento			<> tc.DataVencimento
				OR	tcba.ValorServico			<> tc.ValorServico
				OR	tcba.ValorDescontoNaoCondicionado	<> tc.ValorDescontoNaoCondicionado
				OR	tcba.SomaValorDescontoCondicionado	<> tdc.SomaValorDescontoCondicionado
				OR	tcba.CPFCNPJ				<> r.CPFCNPJ
			)



-- EXEC SPCTT
DROP TABLE #TempTable
DROP TABLE #Parametro
 
SET QUOTED_IDENTIFIER OFF
SET DATEFORMAT dmy
SET LANGUAGE Portuguese
SET CONCAT_NULL_YIELDS_NULL ON

CREATE TABLE #TempTable ( IdUsuario INT, IdUnidade INT, StSistemaEnsinoSuperior BIT, StExecutando_ExcluirDescontoDuplicado BIT NOT NULL DEFAULT (0), StExecutando_LancamentoCobranca_RenumerarRPS BIT NOT NULL DEFAULT (0), StExecutando_LancamentoCobranca_EfetuandoNegociacao BIT NOT NULL DEFAULT (0) )

INSERT #TempTable
	SELECT	IdUsuario = ( SELECT IdUsuario FROM TbUsuario WHERE Login='suporte' ),
		IdUnidade = ( SELECT TOP 1 IdUnidade FROM TbUnidade ),
		StSistemaEnsinoSuperior = 0,
		StExecutando_ExcluirDescontoDuplicado = 0,
		StExecutando_LancamentoCobranca_RenumerarRPS = 0,
		StExecutando_LancamentoCobranca_EfetuandoNegociacao = 0

-- SELECT IdUnidade, NomeUnidade, SiglaUnidade FROM TbUnidade

CREATE TABLE #Parametro ( Parametro VARCHAR(255), ValorVarchar VARCHAR(255), ValorInteger INT, ValorNumeric NUMERIC(12,2), ValorBit BIT )




/*
	Restaurando linha digitável anterior ao bug
*/

UPDATE	tcb

SET		BOL_LinhaDigitavel = tcba.BOL_LinhaDigitavel

FROM	TbTituloCobrancaBoleto tcb
INNER	JOIN VwTituloCobranca tc ON tc.IdTituloCobranca = tcb.IdTituloCobranca
INNER	JOIN TbFormaRecebimento fr ON fr.IdFormaRecebimento = tc.IdFormaRecebimento
INNER	JOIN VPS7_TituloCobrancaBoleto.._CRM229504_OPCAO_TbTituloCobrancaBoleto_Anterior tcba ON tcba.IdTituloCobranca = tcb.IdTituloCobranca

WHERE	fr.BOL_TipoCarteira = 'PJBank_1'
AND	tc.SituacaoTituloCobranca = 'ABE'
AND	tc.SituacaoCobrancaRegistrada IN ('RegistroConfirmado', 'RegistroPendente')
AND	tcba.StRealizarUpdate = 1
```
